### Extracting info from JSON
 We can parse json by converting it into a dictionary.
 This can be done by using the code: **dic = json.loads(x)**

### Extracting text
Referring to the above json, to extract text, we need to use the dictionary and refer to the **text** key.
i.e **dic['text']**

### Extracting date and time of the created tweet
*dic['created_at']*

### Number of favourites
*dic['favorite_count']*

### Retweet count
*dic['retweet_count']*

### Number of images
The json sometimes has a key known as **extended_entities**. If this key exists, that means some kind of media exists in the tweet, it may be video, audio or images.<br/>
This extended entity key is a dictionary which contains a key called **media** which contains info. about the media in the tweet.<br/>
**dic["extended_entities"]['media']**<br/>
media is a list which contains info. about the media.<br/>

Our task is to check whether media contains images, if yes- we have to count the number, if no, we have to return None

```python
#Creating the dataframe
columns = ['text','date and time','favourites','no. of retweets', 'no. of images']
df = pd.DataFrame(columns=columns)
df.head()
```

```python

#Parsing the jsonline file and obtaining the required info

#empty lists for the entries
text = []
dateAndTime = []
fav = []
retweets = []
noImg = []

with jsonlines.open('out.jsonl') as reader:
    for obj in reader:
        dic = json.loads(obj) #loading json and converting it to a dictionary
        
        text.append(dic['text']) #storing the text of tweet
        dateAndTime.append(dic['created_at']) #storing date and time of tweet
        fav.append(dic['favorite_count']) #storing the num of favourites
        retweets.append(dic['retweet_count']) #storing the number of retweets
        c = 0
        if 'extended_entities' in dic.keys():
            d = dic['extended_entities']
            if 'media' in d.keys():
                media_d = d['media']
                #print(d)
                for ob in media_d:
                    if ob['type'] == 'photo':
                    #print('yes')
                        c+=1
        else:
            c = None
        noImg.append(c) #storing the number of images
        
```

Now that we have stored all the information in lists, we can append it to the dataframe created above and obtain the answer

```python
columns = ['text','date and time','favourites','no. of retweets', 'no. of images']
df['text'] = text
df['date and time'] = dateAndTime
df['favourites'] = fav
df['no. of retweets'] = retweets
df['no. of images'] = noImg
```
